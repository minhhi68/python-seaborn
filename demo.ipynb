```python
# Tạo cột `Age_group` với các nhóm dựa trên `Age`
# (0-12]: 'Trẻ em'
# (12-19]: 'Thanh thiếu niên'
# (19-60]: 'Người lớn'
# (60-inf]:  'Người cao tuổi'


titanic['Age_group'] = pd.cut(titanic['Age'], bins=[0, 12, 19, 60, float('inf')],
                        labels=['Trẻ em', 'Thanh thiếu niên', 'Người lớn', 'Người cao tuổi'],
                        right= True) # Nếu right=True (mặc định), giá trị biên bên phải được bao gồm trong khoảng, tức là (a,b]
                                     # Nếu right=False, giá trị biên bên trái được bao gồm trong khoảng, tức là [a,b)

print(titanic['Age_group'].value_counts())

print(f"Số lượng giá trị thiếu: {titanic['Age_group'].isnull().sum()}")
```

```python
# Thay thế các giá trị NaN bằng các giá trị của nhóm tuổi xuất hiện nhiều nhất: Người lớn (từ 19 - 60 tuổi)
# Thay thế giá trị NaN trong cột Age_group thành giá trị "Người lớn"


titanic['Age'] = titanic['Age'].fillna(
    pd.Series(np.random.randint(19, 60, size=len(titanic)))
)

# Thay thế giá trị NaN trong cột Age_group thành giá trị "Người lớn"
titanic['Age_group'] = titanic['Age_group'].fillna("Người lớn")

print(titanic['Age_group'].value_counts())
print()
print(f"Số lượng giá trị thiếu: {titanic['Age'].isnull().sum()}")
```

```python
# Thống kê mô tả cho các cột dữ liệu số
titanic.describe()
```

```python
# Cập nhật giá trị trong cột Survived
# 1: "Sống sót"
# 0: "Tử vong"






titanic['Survived'] = titanic['Survived'].map(lambda x: "Tử vong" if x == 0 else "Sống sót")
```

```python
titanic.head()
```

```python
# Tạo mới cột Class
# Pclass=1: "First"
# Pclass=2: "Second"
# Pclass=3: "Third"







titanic['Class'] = titanic['Pclass'].map(lambda x: "First" if x == 1 else "Second" if x == 2 else "Third")
```

```python
titanic.head()
```

```python
# Tạo thuộc tính 'FamilySize' là kích thước gia đình trên tàu
# FamilySize = SibSp + Parch + 1


titanic['FamilySize'] = titanic['SibSp'] + titanic['Parch'] + 1
titanic[['SibSp', 'Parch', 'FamilySize']].head()
```

```python
bins = [0, 10, 25, 50, 100, 200, 513]
titanic['FareGroup'] = pd.cut(titanic['Fare'], bins=bins, right=False)
titanic['FareGroup'].value_counts()
```

```python
# Tạo nhóm giá vé từ cột 'Fare'
bins = [0, 25, 50, 100, 200, 513]
labels = ['0-dưới 25', '25-dưới 50', '50-dưới 100', '100-dưới 200', '200 trở lên']



# Nếu right=True (mặc định), giá trị biên bên phải được bao gồm trong khoảng, tức là (a,b]
# Nếu right=False, giá trị biên bên trái được bao gồm trong khoảng, tức là [a,b)
titanic['FareGroup'] = pd.cut(titanic['Fare'], bins=bins, labels=labels, right=False)
titanic[['Fare', 'FareGroup']].head()
```

```python
# Tỷ lệ hành khách Sống sót và Tử vong theo Sex (giới tính)





# Tính toán số lượng Sống sót và Tử vong theo Giới tính
count_df = titanic.groupby(['Survived', 'Sex']).size().reset_index(name='No. of passengers')

# Tính toán tỷ lệ
total_per_sex = count_df.groupby('Sex')['No. of passengers'].transform('sum')
count_df['Percentage'] = count_df['No. of passengers'] / total_per_sex * 100

# Vẽ biểu đồ
ax = sns.barplot(data=count_df, x='Sex', y='Percentage', hue='Survived')
for container in ax.containers:
    ax.bar_label(container, fmt='%.1f%%', fontsize=10)

plt.title('Tỷ lệ hành khách Sống sót và Tử vong theo giới tính')
plt.ylabel('Tỷ lệ (%)')
plt.xlabel('Giới tính')

plt.show()
```

```python
#thống kê số lương hành khách theo Survived và Pclass (hạng vé)

plt.figure(figsize=(10,5))
ax = sns.countplot(data=titanic, x = 'Pclass', hue='Survived')
ax.bar_label(ax.containers[0], fontsize=10, fmt ='{:.0f}' )
ax.bar_label(ax.containers[1], fontsize=10, fmt ='{:.0f}' )
plt.title('Số lượng hành khách theo Survived và Pclass (hạng vé)')
plt.xlabel('Hạng vé')
plt.ylabel('Số lượng hành khách')
plt.show()
```

```python
#Tính toán phần trăm tử vong và phần trăm sống sót của các hạng vé

#Tính toán số lượng sống sót và tử vong theo hạng vé
count_df = titanic.groupby(['Pclass', 'Survived']).size().reset_index(name='No. of passengers')

#Tính toán tỷ lệ 
total_per_survived = count_df.groupby('Pclass')['No. of passengers'].transform('sum')
count_df['Percentage'] = count_df['No. of passengers'] / total_per_survived * 100

#Vẽ biểu đò

ax = sns.barplot(data=count_df, x='Pclass', y='Percentage', hue='Survived')
for container in ax.containers:
    ax.bar_label(container, fmt='%.1f%%', fontsize=10)

plt.title('Tỷ lệ hành khách Sống sót và Tử vong theo hạng vé')
plt.ylabel('Tỷ lệ (%)')
plt.xlabel('Hạng vé')
plt.show()
```

```python
#thống kê số lương hành khách theo Survived và Pclass (hạng vé)

plt.figure(figsize=(10,5))
ax = sns.countplot(data=titanic, x = 'Pclass', hue='Survived')
ax.bar_label(ax.containers[0], fontsize=10, fmt ='{:.0f}' )
ax.bar_label(ax.containers[1], fontsize=10, fmt ='{:.0f}' )
plt.title('Số lượng hành khách theo Survived và Pclass (hạng vé)')
plt.xlabel('Hạng vé')
plt.ylabel('Số lượng hành khách')
plt.show()
```

```python
# Tính toán phần trăm tử vong và sống sót théo nhóm giá vé

# Tính toán số lượng Sống sót và Tử vong theo nhóm giá vé
count_df = titanic.groupby(['Survived', 'FareGroup']).size().reset_index(name='Count')

# Tính toán tỷ lệ
total_per_survived = count_df.groupby('FareGroup')['Count'].transform('sum')
count_df['Percentage'] = count_df['Count'] / total_per_survived * 100

# Vẽ biểu đồ
plt.figure(figsize=(10,5))
ax = sns.barplot(data=count_df, x='FareGroup', y='Percentage', hue='Survived', \
                 order = ['0-dưới 25','25-dưới 50','50-dưới 100','100-dưới 200','200 trở lên'])
for container in ax.containers:
    ax.bar_label(container, fmt='%.1f%%', fontsize=10)

plt.title('Tỷ lệ hành khách Sống sót và Tử vong theo Giá vé')
plt.ylabel('Tỷ lệ (%)')
plt.xlabel('Nhóm giá vé')
plt.legend(loc='center right')

plt.show()
```
